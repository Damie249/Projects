```python
# What is the Total Sales Amount and Profit Generated per product category
```

```python
# Group by 'Product_Category' and sum 'Sales_Amount' and 'Profit'
sales_profit_category = transaction_data.groupby("Product_Category")[["Sales_Amount", "Profit"]].sum()


print(f"The Total Sales and Profit per Product Category: \n{sales_profit_category}")
```

```python
# Display how total sales amount relates to customers
```

```python
# Group by 'Customer_ID' and sum 'Sales_Amount'
total_sales_per_customer = transaction_data.groupby("Customer_ID")["Sales_Amount"].sum().sort_values(ascending=False).head(5)

print(f"The Total Sales Amount per Customer: \n{total_sales_per_customer}")
```

```python
# What is the average sales amount per region


# Merge both the customer and transaction table

merged_data = pd.merge(transaction_data, customer_data, on="Customer_ID")
```

```python
merged_data
```

```python
# Group by 'Region' and calculate the average 'Sales_Amount'
average_sales_per_region = merged_data.groupby("Region")["Sales_Amount"].mean().sort_values(ascending=False)
print(f"The Average Sales Amount per Region: \n{average_sales_per_region}")
```

MULTIVARIATE

Regional Trends of the Sales by Product

```python
# Group by 'Product_Category' and 'Region', and then sum the 'Sales_Amount'
sales_by_region = merged_data.groupby(['Product_Category', 'Region'])["Sales_Amount"].sum().unstack()

print(sales_by_region)
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(sales_by_region, annot=True, fmt='.2f')

plt.title("Regional Trend of Sales by Product")
plt.show()
```

```python
# Group by 'Customer_ID' and sum 'Sales_Amount'
customer_sales = transaction_data.groupby("Customer_ID")["Sales_Amount"].sum().sort_values(ascending=False)

customer_sales
```

```python
# How do age, Income and total sales interact

# merge the total sales by customer to the customer table on the customer ID to capture the Age column
table_show = pd.merge(customer_sales, customer_data, on="Customer_ID")[["Customer_ID", "Age", "Income", "Sales_Amount"]]


table_show
```

```python
# create the visuialization

sns.pairplot(table_show)
plt.show()
```

```python
# How do region, age and profit per customer relate


customer_profit = transaction_data.groupby("Customer_ID")["Profit"].sum()
customer_profit
```

```python
customer_age_region = pd.merge(customer_profit, customer_data[["Customer_ID", "Age", "Region"]], on="Customer_ID")

customer_age_region
```

```python
sns.pairplot(customer_age_region, hue="Region")

plt.show()
```
