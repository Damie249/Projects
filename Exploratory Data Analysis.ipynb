Data Dictionary

Here's a brief overview of the dataset you'll be working with:

1. Income: Annual income of the customer(in thousands of dollars)
2. Limit: Credit limit assigned to the customer (in dollars)
3. Rating: Credit rating of the customer
4. Cards: Number of credit cards the customer holds
5. Age: Age of the customer (in years).
6. Education: years of education completed by the customer.
7. Own: Whether the customer owns a home (Yes or No).
8. Student: Whether the customer is a student (Yes or No).
9. Married: Marital status of the customer (Yes or No).
10. Region: Region where the customer resides (South,West, etc)
11. Balance: Balance on the credit card (in dollars)

<b>Income vs. Credit Limit Analysis</b>

* Investigate how customer's income levels influence their assigned credit limits

```python
plt.figure(figsize=(6,4)) # this shows positive relationship, as customer income increases

sns.scatterplot(x="Income",y="Limit", data =df)
plt.xlabel("Income")
plt.ylabel("Credit Limit")
plt.title("Income vs Credit Limit")

plt.show()                        # these customers are low income customers
```

```python
plt.figure(figsize=(6,4)) 

sns.lineplot(x="Income",y="Limit", data =df)
plt.xlabel("Income")
plt.ylabel("Credit Limit")
plt.title("Income vs Credit Limit")

plt.show() 
```

```python
correlation = df["Income"].corr(df["Limit"])
print(f"The correlation between Income and Limit is: {correlation}")# its between -1 and 1


# if it;s negative it means as one is decreasing, one is increasing, in this case it is positve
```

##### <b>Insight</b>:

* Positive correlation between income and credit limit
* Most customers earn 20,000-80,000 with credit limits 2,000-8,000.
* Outliers include low-income individuals with crediyt limits over 11,000


<b> Recommendation</b>

* Tailor credit policies by income segment.
  + High income: Offer premium products with higher limits
  + Lower-to-middle_income: Manage credit limits to balance needs and risk
* Regularly review credit limits with updated income data
* Develop targeted marketing for different income segments.
* Monitor outliers with high credit limits to manage risks
* Implement granular scoring models for better accuracy

<b>* Income, Credit Rating, and Credit Limit:</b> Investigate the combined effect of income and credit raing on credit limit.


<b>Bubble Chart</b>

```python
plt.figure(figsize=(8,6))

sns.scatterplot(x="Income", y="Limit", size="Rating", data =df,alpha=0.2,sizes=(50,500))
plt.xlabel("Income")
plt.ylabel("Limit")
plt.title("Income vs Limit vs Rating")

plt.show()
```

<b>Insight:

*<b>Positive Relationship</b>: There is a clear positive relationship between income and credit limit.

*<b>Clustered Points</b>: Most data points are clsutered in the lower-left corner, with incomes between $20,000 and $60,000 and credit limits between $2,000 and $8000.

*<b> Bubble Sizes</b>: Larger bubbles(higher credit ratings)are located in the top-right indicating that higher incomes and credit limits are associated with better credit ratings.



<b>Recommendation:</b>

* <b>Tailored Credit Policies:</b> Develop differentiated credit policies for customers in the $20,000 to $60,000 income range with credit limits of $2,000 to $8,000, focusing on managing risk and providing credit-building opportunities.
* <b>Premium Products for High Earners</b>: Offer Premium credit products with higher limits and additional benefits to customers with higher incomes and better credit ratings.
* <b>Targeted Financial Educational</b>: Provide financial education and credit improvement programs aimed at customers in the lower income and credit limit ranges to help them improve their credit scores and financial health.
* <b>Leverage Clustering</b>: Use the dense clustering of data points to design targeted marketing campaigns and personalised financial product needs of these income segments.
* <b>Monitor and Support</b>: Regularly monitor customers with higher credit limits and incomes for early signs of financial distress and provide proactive support to maintain their credit health.
