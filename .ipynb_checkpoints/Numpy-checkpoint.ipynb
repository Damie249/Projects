#### Introduction to Pandas

* Pandas is a powerful and widely used open source data manipulation and analysis library for python
* Developed on top of the NumPy Library, pandas introduces two primary data structures: the Series(1D) and Dataframe(2D)
* These structures provide intuitive and efficient tools for handling and analysing structured data.

Key Features:

1. Dataframe: The DataFrame is a two-dimensional, tabular data structure that resemblesa spreadsheet or DQL, table. It allows for easy imdexing, filtering and manipulation of data.
2.  Series: The series is one-dimensional tabular array, providing a flexible and efficient structure for handling time series data, numerical arrays or any other single dimensional dataset.
3.  Data Alignment: Pandas simplifies data alignment and integration, enabling users to merge join and concatenate datasets with ease.
4.  Label-based indexing: With Pandas, you can label rows and columns, making it intuitive to access and manipulate specific data points within Dataframe.
5.  Missing Data Handling: Pandas provides robust tools for dealing with missing and incomplete data, ensuring a smooth data analysis process.
6.  Time Series and Data Functionalities: Pandas excels in handling time series data, offering functinality for data range generation, frequency conversion and time based indexing
7.  Wide Range of Operations: From grouping and aggregating data to reshaping and pivoting. Pandas supports a vast array of operations essential for comprehensive data analysis.
8.  Integration with Other Libraries: Pandas seamlessly integrates with visualization libraies like Matplotlib and statistical libraries like SciPy, creating a cohesive environment for data exploration and analysis.

Whether you are a beginner exploring the basics of data analysis or an experienced data scientist working on complex projects, Pandas is an invaluable tool that streamlines the process of data manipulation and analysis as Python. You can always refer to the Pandas documentaion for more information.

Merge

```python
customers = pd.DataFrame({"CustomerID": [100, 101, 103],
                        "CustomerName": ["Ade", "Usman", "Sarah"]
                        })

orders = pd.DataFrame({"CustomerID": [100, 100, 102, 104],
                    "OrderDate": ["2023-06-15", "2023-06-20", "2023-06-19", "2023-06-22"]
                        })
customers
```

```python
orders
```

```python
pd.concat([customers,orders],ignore_index=True)
```

```python
# Inner Merge

pd.merge(customers, orders, on= "CustomerID", how="inner")
```

```python
# Left Merge

pd.merge(customers, orders, on= "CustomerID", how="left")
```

```python
# Right Merge

pd.merge(customers, orders, on= "CustomerID", how="right")
```

```python
# Outer Merge -is the concatenation

pd.merge(customers, orders, on= "CustomerID", how="outer")
```

Pivot

```python
data = {"Name": ["Ade","Obinna", "Usman" ,"Sarah", "Flora", "Ben"],
     "Age": [12,15,14,28,13,21],
     "City": ["New York", "Abuja", "Lagos", "Kano", "Abuja", "Jos"],
    "Department": ["HR", "Finance", "IT", "Marketing", "IT", "HR"],
    "YearsOfExperience": [ 5,6,8,5,7,8],
    "Salary": [7000,8500,5000, 9000, 7500, 8000]
        
    }
df = pd.DataFrame(data)
df
```

```python
df.pivot(index="City", columns="Department", values="Salary")
```

```python
df.pivot(index="City",columns="Department", values= ["Name","Salary"])
```

```python
df.pivot_table(index= "City", columns = "Department", values=["Salary"])
```

```python
df.pivot_table(index= "City", columns = "Department", values=["Salary"], aggfunc=["sum"])
```

```python
df.pivot_table(index= "City", columns = "Department", values=["Salary"], aggfunc=["sum","mean"])
```

```python
df.pivot_table(index=["City"],columns="Department",values="Salary",aggfunc=["sum", "mean"],fill_value="No Record")
```
