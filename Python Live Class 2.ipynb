```python
# Top 5 products driving the highest revenue for the company


# Select top 5,
#sum(total sales) as Revenue
#from sales data
#Groupby Product name
#Order by Revenue Desc



#Table name = sales_data
# Sum(Total Sales)= Revenue
#Column name = Product Name
#Column to be added = Total Sales
#Group by = Product Name
#Order by = Revenue(Total Sales)
```

```python
# Question 1: Top 5 products driving the highest revenue for the company

top_products = sales_data.groupby("Product Name")["Total Sales"].sum().sort_values(ascending=False).head(5)
top_products

# step 1- create a variable name(top products)
# 2. Call the table that includes the data (sales_data)
# 3. Group the data by the required output column(Product name)
# 4 Sort the values (.sort_values.......)
```

```python
# Question 2 : products in danger of going out of stock

#Select Product Name, Sum (Stock) As Stock_Value
#From Products
#Where Stock_Value < 20
#Group by Product Name

low_products_stock = products_data[products_data["Stock"] < 20]

print("Products likely to go out of stock soon: \n", low_products_stock[["Product Name", "Stock"]])
```

```python
# Question 3: Average Purchase 

#Select Region, AVG(Total Sales) As Avg_Total
#From Sales_data
#Group by Region


sales_data["Region"]
```

```python
# Python Code
average_sales_by_region = sales_data.groupby("Region")["Total Sales"].mean()
average_sales_by_region
```

```python
# Question 4: Customer Segment purchasing the most expensive products

customer_segment = sales_data[sales_data["Price"]> 1000]
customer_segment
```

```python
# Sales by Region

sales_by_region = sales_data.groupby(["Product Name", "Region"])["Total Sales"].sum().unstack()
sales_by_region
```

```python
# Visualizations


#Heat map, Line Chart, Histogram

# Import necessary libraries for visualizations

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visual 1: Regional Sales Trend

# Sales by product by Region using a Heatmap

plt.figure(figsize=(5, 6))
sns.heatmap(sales_by_region, annot=True, fmt='.2f', cmap="Blues")
plt.show()

```

```python
# bar chart of total sales by product

plt.figure(figsize= (30,8))
plt.bar(sales_data["Product Name"], sales_data["Total Sales"])
plt.title("Total Sales by Product")
```

```python
# Line plot to visualize sales trend

sales_trends = sales_data.groupby("Product Name")["Total Sales"].sum()
sales_trends.plot (kind = "line")

plt.show()
```
