### Feature Engineering

#### 1. Recency

```python
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"])
```

```python
# data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"], dayfirst=True, errors="coerce")
```

```python
data.info()
```

```python
data["InvoiceDate"]
```

```python
max_date = max(data["InvoiceDate"])
```

```python
max_date
```

```python
data["Diff"] = max_date - data ["InvoiceDate"]
```

```python
data.head()
```

```python
data["Diff"] = data["Diff"].dt.days
```

```python
data.head()
```

```python
recency = data.groupby("CustomerID")["Diff"].min()
```

```python
recency = recency.reset_index()
```

```python
recency
```

### 2. Frequency

```python
data.head()
```

```python
frequency = data.groupby("CustomerID")["InvoiceNo"].count()
```

```python
frequency
```

```python
frequency = frequency.reset_index()
```

```python
frequency
```

```python
frequency.columns = ["CustomerID", "Frequency"] # to change the name of the column to frequency instead of invoice no
```

```python
frequency
```

### 3. Amount

```python
data["Amount"] = data["Quantity"] * data["UnitPrice"]
```

```python
data.head()
```

```python
amount = data.groupby("CustomerID")["Amount"].sum()
amount = amount.reset_index()
```

```python
amount
```

###  Merge all 3 into a single data frame

```python
rfa = pd.merge(recency, frequency, how="inner",on="CustomerID")
```

```python
rfa = pd.merge(rfa, amount, how="inner",on="CustomerID")
```

### Modelling

K-means clustering is one of the simplest and popular unsupervised machine learning algorithms.

The algorithm works as follows:

* First we initialize k points, called means randomly
* We categorize each item to its closest mean and we update the mean's coordinates, which are the averages of the items categorized in that mean so far.
* We repeat the process for a given number of iterations and at the end, we have our clusters.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, n_init=10, max_iter=50)
```

```python
kmeans.fit(rfa_df_scaled)
```

```python
rfa_df_scaled.head()
```

```python
kmeans.labels_
```

```python
kmeans.inertia_ # or how close the datapoints are to each other, this is th eeror, you want to minimize it as much as possible
```

```python
# when you overfit, the model will just cram and not learn, so avoid this

# dont also underfit
```

### Select the optimum number of clusters
 Elbow curve to get the right number of clusters
A fundamental step for any unsupervised unsupervised algorithm the optimal number of clusters into which the data may be clustered

```python
Image(filename='Elbow.png', width=300)
```

```python
# your elbow joint is what you are looking for., here it is 4
```

```python
# within cluster sum of cells 
wcss = []

for num_clusters in range(1,11):
   kmeans = KMeans(n_clusters=num_clusters, n_init=10, max_iter=50)
   kmeans.fit(rfa_df_scaled)
   wcss.append(kmeans.inertia_)
```

```python
plt.plot(wcss)
plt.xticks(list(range(1,11)))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# we can pick 2 or 3 as elbow joint===hmm since we dont have the same dataset
```

```python
kmeans = KMeans(n_clusters=3, n_init=10, max_iter=50)
kmeans.fit(rfa_df_scaled)# use 3 clusters
```

```python
kmeans.labels_
```

```python
rfa["Category"] = kmeans.labels_
```

```python
rfa.head()
```
