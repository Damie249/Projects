```python
### Handle Outliers - in bmi and charges, it is clear there is no lower bound outliers
```

```python
Q1 = df[["bmi","charges"]].quantile(0.25)
Q3 = df[["bmi", "charges"]].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
```

```python
lower_bound # anything less than the below is an outlier
```

```python
upper_bound # anything above this is an outlier
```

```python
# the below will remove the outlier
```

```python
numerical_data = numerical_data[(numerical_data["bmi"]<=upper_bound["bmi"]) & (numerical_data["charges"] <=upper_bound["charges"])]
```

### Data Pre-processing

```python
target = df["charges"]
```

```python
features = df.drop("charges", axis=1)
```

```python
features.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
categorical_data = features.select_dtypes(include="object")
```

```python
categorical_data.head()
```

```python
for column_name in categorical_data:
    features[column_name] = encoder.fit_transform(features[column_name])
```

```python
features.head()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
for column_name in features:
    features[column_name] = scaler.fit_transform(features[[column_name]])
```

```python
features.head()
```

### Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

### Modelling

```python
from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print (f"MAE: {mae}")
print (f"MSE: {mse}")
print (f"RMSE: {rmse}")
print (f"R2: {r2}")  
sns.regplot(x=y_pred,y=y_test, color="green", marker="+")
```

```python
target
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
print (f"MAE: {mae}")
print (f"MSE: {mse}")
print (f"RMSE: {rmse}")
print (f"R2: {r2}")  
sns.regplot(x=y_pred,y=y_test, color="green", marker="+")
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
print (f"MAE: {mae}")
print (f"MSE: {mse}")
print (f"RMSE: {rmse}")
print (f"R2: {r2}")  
sns.regplot(x=y_pred,y=y_test, color="green", marker="+")
```
